```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from japanmap import picture
```

```
df_con = pd.read_html('https://uub.jp/pdr/m/c.html')
```

```
df_tmp = df_con[4].set_index(1)
df_tmp
```

```
df_tmp.columns = df_tmp.loc['都道府県']
df_tmp
```

```
df_tmp = df_tmp.drop(['＃','合計↓'],axis=1).drop(['都道府県','2020年3月末','▲','全国'],axis=0)
df_tmp
```

```
df_result = df_tmp.replace('-',0).astype(int).idxmax(axis=1)
```

```
df_result.value_counts()
```

```
cmap = {'セブンイレブン':'green','ファミリーマート':'lightblue','ローソン':'blue','セイコーマート':'red'}
fig = plt.subplots(figsize=(10,10))
plt.imshow(picture(df_result.replace(cmap)))
```

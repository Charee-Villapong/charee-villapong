### 第2問
次の表な掛け算の九九表を作成しなさい。
1から9までの数値を繰り返しを用いて実現させなさい。printで九九表のイメージになる文字を出力しただけと言うのは認めません。<br>
プログラムとして実現してください<br>
<img src="img/9x9.png">

```python
for i in range(1, 10):
    for s in range(1, 10):
        kakeru = str(i*s)
        result = kakeru.rjust(2)
        print(result, end=" ")
    print()
```

```python
for i in range(1,10):
    for ii in range(1,10):
        kakeru = i * ii.reshape(9,9)
        print(kakeru)
```

```python
for i in range(1, 10):
    print(i)
    for j in range(1, 10):
        print(j)
        kakeru = str(i * j)
        result = kakeru.rjust(3)
        print(result,end=" ")
    print()
    
```

```python
for i in range(1, 10):
    for j in range(1, 10):
        print(i*j,end=" ")
    print()
```

```
import pandas as pd
from pandas_datareader import data
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```
start_date = '2020-01-01'
end_date = '2021-01-01'

df = data.DataReader('^N225','yahoo',start_date,end_date)
```

```
df.head(5)
```

```
date = df.index
price = df['Adj Close']
```

```
plt.figure(figsize=(30,10))
plt.plot(date,price)
```
